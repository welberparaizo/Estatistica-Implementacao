Parâmetros da distribuição normal

```python
# Parâmetros da distribuição normal
mu = 0      # Média
sigma2 = 2  # Variância

# Resultados
print(f"Média (Esperança) da distribuição N(0, 2): {mu}")
print(f"Variância da distribuição N(0, 2): {sigma2}")
```

Parâmetros da distribuição Uniforme

```python
# Definindo os limites da distribuição
a = -1
b = 1

# Calculando a média
media = (a + b) / 2

# Calculando a variância
variancia = ((b - a) ** 2) / 12

# Resultados
print(f"Média (Esperança) da distribuição U[-1, 1]: {media}")
print(f"Variância da distribuição U[-1, 1]: {variancia}")
```

Parâmetro da distribuição exponencial

```python
# Parâmetro da distribuição exponencial
lambda_param = 0.05

# Calculando a média
media = 1 / lambda_param

# Calculando a variância
variancia = 1 / (lambda_param ** 2)

# Resultados
print(f"Média (Esperança) da distribuição E(λ=0,05): {media}")
print(f"Variância da distribuição E(λ=0,05): {variancia}")
```

Parâmetros da distribuição binomial

```python
# Parâmetros da distribuição binomial
n = 15         # Número de tentativas
p = 0.40       # Probabilidade de sucesso

# Calculando a média
media = n * p

# Calculando a variância
variancia = n * p * (1 - p)

# Resultados
print(f"Média (Esperança) da distribuição B(15, 0,40): {media}")
print(f"Variância da distribuição B(15, 0,40): {variancia}")
```

```python
# Parâmetros iniciais
Tamanho_amostras = [5, 10, 50]
Numero_amostras = 10000

# Dicionário para armazenar resultados
Dicionario = {
    "Média Amostral": {},
    "Variância Amostral": {}
}

# Estatísticas populacionais
media_populacional = data.mean().values[0]
variancia_populacional = data.var(ddof=1).values[0]

# Loop para gerar as amostras
for n in Tamanho_amostras:
    medias = []
    variancias = []
    
    # Gerar amostras aleatórias
    for i in range(Numero_amostras):
        amostra = data.sample(n, replace=True)
        media_amostral = amostra.mean().values[0]
        variancia_amostral = amostra.var(ddof=1).values[0]

        medias.append(media_amostral)
        variancias.append(variancia_amostral)
        
    # Armazenar os resultados
    Dicionario["Média Amostral"][f"{n}"] = medias
    Dicionario["Variância Amostral"][f"{n}"] = variancias

# Definir limites globais para os eixos
x_min_media = min(min(Dicionario["Média Amostral"][f"{n}"]) for n in Tamanho_amostras)
x_max_media = max(max(Dicionario["Média Amostral"][f"{n}"]) for n in Tamanho_amostras)
x_min_var = min(min(Dicionario["Variância Amostral"][f"{n}"]) for n in Tamanho_amostras)
x_max_var = max(max(Dicionario["Variância Amostral"][f"{n}"]) for n in Tamanho_amostras)

# Gerar histogramas e curvas teóricas para cada n
for n in Tamanho_amostras:
    # Parâmetros teóricos para a curva normal das médias amostrais
    media_teorica = media_populacional
    desvio_teorico_media = np.sqrt(variancia_populacional / n)

    # Parâmetros para a curva das variâncias amostrais
    media_teorica_var = variancia_populacional
    desvio_teorico_var = np.sqrt(2 * (variancia_populacional ** 2) / (n - 1))
    
    # Parâmetros para a curva qui-quadrado das variâncias amostrais
    k = n - 1  # graus de liberdade
    escala = variancia_populacional / k  # escala para ajustar o qui-quadrado à variância amostral
    
    plt.figure(figsize=(12, 5))
    
    # Histograma das médias amostrais com curva normal teórica
    plt.subplot(1, 2, 1)
    plt.hist(Dicionario["Média Amostral"][f"{n}"], bins=30, color='red', edgecolor='black', alpha=0.4, density=True, label="Histograma")
    x = np.linspace(x_min_media, x_max_media, 1000)
    y = norm.pdf(x, media_teorica, desvio_teorico_media)
    plt.plot(x, y, 'b--', label="Curva Normal Teórica")
    plt.title(f"Histograma da Média Amostral com Curva Teórica (n={n})")
    plt.xlabel("Média Amostral")
    plt.ylabel("Densidade")
    plt.xlim(x_min_media, x_max_media)
    plt.legend()
    
    # Histograma das variâncias amostrais com curva qui-quadrado teórica
    plt.subplot(1, 2, 2)
    plt.hist(Dicionario["Variância Amostral"][f"{n}"], bins=30, color='purple', edgecolor='black', alpha=0.4, density=True, label="Histograma")
    x = np.linspace(x_min_var, x_max_var, 1000)
    y = chi2.pdf(x / escala, df=k) / escala  # Ajuste da escala da qui-quadrado
    plt.plot(x, y, 'g--', label="Curva Qui-Quadrado Teórica")
    plt.title(f"Histograma da Variância Amostral com Curva Qui-Quadrado (n={n})")
    plt.xlabel("Variância Amostral")
    plt.ylabel("Densidade")
    plt.xlim(x_min_var, x_max_var)
    plt.legend()
    
    plt.tight_layout()
    plt.show()
    
for key in Dicionario["Média Amostral"]:
    print(f"Média amostral de tamanho amostra aleatoria n={key}: {np.mean(Dicionario['Média Amostral'][key]):.4f}")

for key in Dicionario["Variância Amostral"]:
    print(f"Variância amostral de tamanho amostra aleatoria n={key}: {np.mean(Dicionario['Variância Amostral'][key]):.4f}")
```
